```python
import easyocr
reader = easyocr.Reader(['en', 'hi'])
```

```python
result = reader.readtext('/content/rsign.jpg',paragraph=True)
result
```

```python
import cv2
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX

```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import ImageFont, ImageDraw, Image
```

```python
img = cv2.imread('/content/rsign.jpg')
spacer = 300
font = cv2.FONT_HERSHEY_SIMPLEX
for detection in result:
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    #img = cv2.putText(img,text,(20,spacer), font, 0.5,(255,255,0),2,cv2.LINE_AA)
    spacer+=15
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```

```python
result = reader.readtext('aadhar.jpg',paragraph=True)
result
```

```python
img = cv2.imread('aadhar.jpg')
spacer = 100
for detection in result:
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```
